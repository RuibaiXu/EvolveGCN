```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import torch
import pandas as pd
reports = torch.load('./data/reports.pt')
loss = [r['loss'] for r in reports]
acc = [r['acc'] for r in reports]
```

```python
x = np.array(list(range(len(reports))))
y_loss = loss
y_acc = acc
loss_plt = plt.subplot(1, 2, 1)
plt.plot(x, y_loss, label='loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()
acc_plt = plt.subplot(1, 2, 2)
plt.plot(x, y_acc, label='acc', color='red')
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
